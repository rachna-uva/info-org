```python
import pandas as pd
import os
import re
```

```python
df_2018 = pd.read_csv("data/arabica_data_cleaned_v2.csv")  
''' v2 Updates:
ID 1198 updated to country "colombia", as farms by the same owner are located there
ID 1312 was removed as they score zero on everything
'''
df_2023 = pd.read_csv("data/2023_coffee-quality-data-CQI/df_arabica_clean.csv")
```

```python
display(df_2018.head())
display(df_2023.head())
```

```python
# Define columns to keep
columns_to_keep = ['Species', 'Country.of.Origin', 'Harvest.Year', 'Color', 'Processing.Method', 'Aroma', 
                  'Flavor', 'Aftertaste', 'Acidity', 'Body', 'Balance', 
                  'Uniformity', 'Clean.Cup', 'Sweetness', 'Cupper.Points', 
                  'Total.Cup.Points']

# Filter 2018 dataframe
df_2018_filtered = df_2018[columns_to_keep].copy()

# Create mapping for 2023 column names (spaces instead of dots)
column_mapping_2023 = {
    'Country of Origin': 'Country.of.Origin',
    'Processing Method': 'Processing.Method',
    'Clean Cup': 'Clean.Cup',
    'Overall': 'Cupper.Points',  # Overall maps to Cupper.Points
    'Total Cup Points': 'Total.Cup.Points',
    'Harvest Year': 'Harvest.Year'
}

# Rename 2023 columns to match 2018 format
df_2023_renamed = df_2023.rename(columns=column_mapping_2023)

# Add Species column to 2023 data (all Arabica)
df_2023_renamed['Species'] = 'Arabica'

# Filter 2023 dataframe to keep only the same columns
df_2023_filtered = df_2023_renamed[columns_to_keep].copy()
```

```python
# Display the results
print("2018 Filtered DataFrame:")
display(df_2018_filtered.head())
print(f"\nShape: {df_2018_filtered.shape}")

print("\n" + "="*50 + "\n")

print("2023 Filtered DataFrame:")
display(df_2023_filtered.head())
print(f"\nShape: {df_2023_filtered.shape}")

# Check if all columns match
print("\n" + "="*50 + "\n")
print("Column alignment check:")
print(f"2018 columns: {list(df_2018_filtered.columns)}")
print(f"2023 columns: {list(df_2023_filtered.columns)}")
print(f"Columns match: {list(df_2018_filtered.columns) == list(df_2023_filtered.columns)}")
```

```python
def clean_harvest_year(year_str):
    """Simple cleaning: extract years, return single/range/unknown."""
    if pd.isna(year_str):
        return 'Unknown'
    
    year_str = str(year_str).strip()
    
    # Find all 4-digit years between 2000-2025
    years = re.findall(r'\b(20[0-2][0-9])\b', year_str)
    years = sorted(set(years))
    
    if len(years) == 1:
        return years[0]
    elif len(years) == 2:
        return f"{years[0]}-{years[1]}"
    elif len(years) > 2:
        raise ValueError(f"Found more than 2 years in: '{year_str}' - years found: {years}")
    
    # Handle special cases
    special_cases = {
        '4T/10': '2010',
        '4T72010': '2010',
        '08/09 crop': '2008-2009',
        
    }
    
    if year_str in special_cases:
        return special_cases[year_str]
    
    return 'Unknown'
```

```python
# Apply cleaning
df_2018_filtered['Harvest.Year.Original'] = df_2018_filtered['Harvest.Year']
df_2018_filtered['Harvest.Year'] = df_2018_filtered['Harvest.Year'].apply(clean_harvest_year)

df_2023_filtered['Harvest.Year.Original'] = df_2023_filtered['Harvest.Year']
df_2023_filtered['Harvest.Year'] = df_2023_filtered['Harvest.Year'].apply(clean_harvest_year)

# Show results
print("2018 Dataset - Sample Transformations:")
display(df_2018_filtered[['Harvest.Year.Original', 'Harvest.Year']].drop_duplicates().head(20))

print("\n2023 Dataset - Sample Transformations:")
display(df_2023_filtered[['Harvest.Year.Original', 'Harvest.Year']].drop_duplicates().head(20))
```

```python
# Fill missing values in Color with 'Unknown'
df_2018_filtered['Color'] = df_2018_filtered['Color'].fillna('Unknown')
df_2023_filtered['Color'] = df_2023_filtered['Color'].fillna('Unknown')

# Map 2023 Processing Methods to match 2018 categories (only map different ones)
processing_method_mapping = {
    'Double Anaerobic Washed': 'Other',
    'Double Carbonic Maceration / Natural': 'Natural / Dry',
    'Wet Hulling': 'Other',
    'Honey,Mossto': 'Pulped natural / honey',
    'Anaerobico 1000h': 'Other',
    'Semi Washed': 'Semi-washed / Semi-pulped',
    'SEMI-LAVADO': 'Semi-washed / Semi-pulped',
}

df_2023_filtered['Processing.Method'] = df_2023_filtered['Processing.Method'].replace(processing_method_mapping)

# Fill missing values in Processing.Method with 'Unknown'
df_2018_filtered['Processing.Method'] = df_2018_filtered['Processing.Method'].fillna('Unknown')
df_2023_filtered['Processing.Method'] = df_2023_filtered['Processing.Method'].fillna('Unknown')
```

```python
def column_stats(df, name):
    print(f"\n{name} Dataset:")
    print("-" * 80)
    for col in df.columns:
        if col.endswith('.Original'):
            continue
        total = len(df)
        non_null = df[col].notna().sum()
        null = df[col].isna().sum()
        unique = df[col].nunique()
        print(f"{col:25} | Total: {total:5} | Non-null: {non_null:5} | Null: {null:5} | Unique: {unique:5}")

column_stats(df_2018_filtered, "2018")
column_stats(df_2023_filtered, "2023")
```

```python
output_folder = 'data/cleaned_coffee_data'

# Drop the .Original columns before saving
df_2018_clean = df_2018_filtered.drop(columns=['Harvest.Year.Original'])
df_2023_clean = df_2023_filtered.drop(columns=['Harvest.Year.Original'])

# Save individual cleaned datasets
df_2018_clean.to_csv(f'{output_folder}/coffee_2018_cleaned.csv', index=False)
df_2023_clean.to_csv(f'{output_folder}/coffee_2023_cleaned.csv', index=False)

# Combine both datasets
df_combined = pd.concat([df_2018_clean, df_2023_clean], ignore_index=True)

# Save combined dataset
df_combined.to_csv(f'{output_folder}/coffee_combined_2018_2023.csv', index=False)
```
