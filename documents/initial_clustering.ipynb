```python
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
```

```python
data = pd.read_csv('merged_data_cleaned.csv')

data = data[data["Unnamed: 0"] != 1310]
```

```python
# Features used in clustering
features = data[['Aroma', 'Flavor', 'Aftertaste', 'Acidity', 'Body', 'Balance',
                 'Uniformity', 'Clean.Cup', 'Sweetness', 'Cupper.Points', 'Total.Cup.Points']]

# Standardize the data
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)
```

```python
# Choose the number of clusters
kmeans = KMeans(n_clusters=2, random_state=40)

# Fit the model
data['Cluster'] = kmeans.fit_predict(scaled_features)
```

```python
# Reduce dimensions to 2D
pca = PCA(n_components=2)
pca_components = pca.fit_transform(scaled_features)

# Plot the clusters
plt.figure(figsize=(8,6))
plt.scatter(pca_components[:, 0], pca_components[:, 1], c=data['Cluster'], cmap='viridis')
plt.title('Cluster Visualization')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.colorbar()
plt.show()
```

```python
# Select only numeric columns for grouping
numeric_columns = data.select_dtypes(include=['float64', 'int64']).columns

# Group by cluster and calculate the mean for only numeric columns
analysis = data.groupby('Cluster')[numeric_columns].mean()

# Display the results
analysis
```

```python
# Count the number of people in each cluster
cluster_counts = data['Cluster'].value_counts()

# Display the count for each cluster
cluster_counts
```

```python
# Create subplots for each feature
features_to_plot = ['Aroma', 'Flavor', 'Aftertaste', 'Acidity', 'Body', 'Balance',
                   'Uniformity', 'Clean.Cup', 'Sweetness', 'Cupper.Points', 'Total.Cup.Points']

# Set up the subplot grid
fig, axes = plt.subplots(4, 3, figsize=(15, 12))
axes = axes.ravel()  # Flatten the 2D array of axes for easy indexing

for i, feature in enumerate(features_to_plot):
    # Create boxplot for each feature by cluster
    data.boxplot(column=feature, by='Cluster', ax=axes[i])
    axes[i].set_title(f'{feature} by Cluster')
    axes[i].set_ylabel(feature)
    axes[i].set_xlabel('Cluster')

# Remove any empty subplots
for i in range(len(features_to_plot), len(axes)):
    fig.delaxes(axes[i])

plt.suptitle('Feature Distributions by Cluster')  # Overall title
plt.tight_layout()
plt.show()

# Alternative: Bar plot showing mean values for each feature by cluster
plt.figure(figsize=(14, 8))
cluster_means = data.groupby('Cluster')[features_to_plot].mean()

# Plot for each cluster
x = range(len(features_to_plot))
width = 0.35  # Width of bars

for cluster in sorted(data['Cluster'].unique()):
    plt.bar([i + width*cluster for i in x], cluster_means.loc[cluster], 
            width=width, label=f'Cluster {cluster}', alpha=0.7)

plt.xlabel('Features')
plt.ylabel('Mean Values')
plt.title('Mean Feature Values by Cluster')
plt.xticks([i + width for i in x], features_to_plot, rotation=45)
plt.legend()
plt.tight_layout()
plt.show()

# Individual feature plots (more detailed)
for feature in features_to_plot:
    plt.figure(figsize=(10, 6))
    
    # Create subplots for each cluster
    for cluster in sorted(data['Cluster'].unique()):
        cluster_data = data[data['Cluster'] == cluster][feature]
        
        # Plot histogram
        plt.hist(cluster_data, alpha=0.6, label=f'Cluster {cluster}', bins=20)
    
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    plt.title(f'Distribution of {feature} by Cluster')
    plt.legend()
    plt.tight_layout()
    plt.show()

# Violin plots for better distribution visualization
plt.figure(figsize=(14, 10))
for i, feature in enumerate(features_to_plot[:6]):  # First 6 features
    plt.subplot(2, 3, i+1)
    
    # Prepare data for violin plot
    plot_data = [data[data['Cluster'] == cluster][feature] for cluster in sorted(data['Cluster'].unique())]
    
    plt.violinplot(plot_data, positions=range(len(plot_data)))
    plt.title(feature)
    plt.xlabel('Cluster')
    plt.xticks(range(len(plot_data)), [f'Cluster {c}' for c in sorted(data['Cluster'].unique())])

plt.suptitle('Feature Distributions by Cluster (Violin Plots)')
plt.tight_layout()
plt.show()

# Radar chart for comparing cluster profiles
from math import pi

# Calculate mean values for each cluster
cluster_means = data.groupby('Cluster')[features_to_plot].mean()

# Number of variables
categories = list(features_to_plot)
N = len(categories)

# Create radar chart
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, polar=True)

# Compute angles for each category
angles = [n / float(N) * 2 * pi for n in range(N)]
angles += angles[:1]  # Complete the circle

# Plot each cluster
for cluster in cluster_means.index:
    values = cluster_means.loc[cluster].values.tolist()
    values += values[:1]  # Complete the circle
    ax.plot(angles, values, linewidth=2, label=f'Cluster {cluster}')
    ax.fill(angles, values, alpha=0.1)

# Add category labels
plt.xticks(angles[:-1], categories, size=8)
plt.yticks(color="grey", size=7)
plt.title('Cluster Profiles - Radar Chart', size=14)
plt.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))
plt.tight_layout()
plt.show()

# Statistical summary by cluster
print("Statistical Summary by Cluster:")
for cluster in sorted(data['Cluster'].unique()):
    print(f"\n--- Cluster {cluster} ---")
    cluster_data = data[data['Cluster'] == cluster][features_to_plot]
    print(f"Number of samples: {len(cluster_data)}")
    print("Mean values:")
    print(cluster_data.mean().round(2))
```
